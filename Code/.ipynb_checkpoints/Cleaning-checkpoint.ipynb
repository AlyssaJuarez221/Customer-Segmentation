## Outliers

```python
#Gather numerial columns
datanum = data.select_dtypes(include=np.number)
datanum.head()
datanum.shape
```

```python
def plot_boxplot(df, feature):
    print(feature)
    df.boxplot(column=[feature])
    plt.grid(False)
    plt.show()
```

```python
for feature in datanum:
    plot_boxplot(datanum, feature)
```

```python
def outliers(df, ft):
    Q1 = df[ft].quantile(0.25)
    Q3 = df[ft].quantile(0.75)
    IQR = Q3 - Q1
    upper = Q3 + 1.5 * IQR
    lower = Q1 - 1.5 * IQR
    ls = df.index[ (df[ft] < lower) | (df[ft] > upper)]
    return ls
```

### I felt it neccesary to keep some of the outlier values in order to prioritize what I am looking to achieve.

```python
index_list = []
for feature in ['Income', 'Age']:
    index_list.extend(outliers(data, feature))
```

```python
def remove(df, ls):
    ls = sorted(set(ls))
    df = df.drop(ls)
    return df
```

```python
df_cleaned = remove(data, index_list)
df_cleaned.shape
```

## Categorical to Numerical

```python
datacat = data.select_dtypes(exclude=["number"])
datacat.head()
```

```python
LE=LabelEncoder()
for i in datacat:
    df_cleaned[i]=df_cleaned[[i]].apply(LE.fit_transform)
```

## PCA

### Scree plot made in R

### From this we can assume 3 is the number of components needed

```python
pca = PCA(n_components=3)
pca.fit(scaled_df)
PC_scores = pd.DataFrame(pca.transform(scaled_df),
               columns = ['PC1', 'PC2', 'PC3'])
PC_scores.describe().T
```
